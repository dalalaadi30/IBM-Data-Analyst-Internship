```python
req_data = df[["country","year","co2","methane","nitrous_oxide"]]
req_data = req_data[req_data["year"] > 1989]

dev_countries = req_data[req_data["country"].isin(["Australia","France","Germany","Switzerland","Norway","Finland","Japan","South Korea","United States","China"])] 
dev_countries.head()
```

```python
developed_ext = df[["country", "year", "co2", "oil_co2", "cement_co2", "gas_co2", "coal_co2", "consumption_co2", "land_use_change_co2", "flaring_co2"]]
developed_ext = developed_ext[developed_ext["country"].isin(["United States", "China", "Germany", "Australia"])]
developed_ext = developed_ext[developed_ext["year"] > 2015]
countries3 = developed_ext['country'].unique()

# Define the number of categories (emission types)
categories = ['oil_co2', 'coal_co2', 'gas_co2', 'consumption_co2', 'cement_co2', 'land_use_change_co2', 'flaring_co2']
n_categories = len(categories)
bar_width = 0.1  # Width of each bar

fig, axes = plt.subplots(len(countries3), 1, figsize=(12, 6 * len(countries3)))

# Plot emissions for each country
for i, country in enumerate(countries3):
    country_data = developed_ext[developed_ext['country'] == country]
    years = country_data['year']
    
    # Adjust positions for each category's bars
    for j, category in enumerate(categories):
        bar_positions = [year + j * bar_width for year in years]
        axes[i].bar(bar_positions, country_data[category], width=bar_width, label=category.replace('_', ' ').title())

    axes[i].set_title(f'Emissions in {country}')
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Emissions')
    axes[i].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

file_path = '/kaggle/working/dev_countries.csv'  
df = pd.read_csv(file_path)


df_filtered = df[(df['year'] >= 2016) & (df['year'] <= 2023)]

color_min = df_filtered['co2'].quantile(0.05)  
color_max = df_filtered['co2'].quantile(0.95)  

fig = px.choropleth(
    df_filtered,
    locations='country',  # Column containing country names
    locationmode='country names',  # Set the location mode to country names
    color='co2',  
    color_continuous_scale='oranges',  
    range_color=(color_min, color_max),  
    animation_frame='year',  # Column containing the years
    title='Choropleth Map of CO2 Emissions by Developed Countries (2016-2023)',
    hover_name='country',  
    projection='natural earth'  
)

fig.update_layout(
    width=1200,  
    height=800,  
    margin=dict(l=0, r=0, t=50, b=0),  # Adjust margins
    title_font_size=24,  # Set title font size
    title_x=0.5,  # Center the title
    geo=dict(
        showframe=False,  
        showcoastlines=True,  
        coastlinecolor="Black",  
        projection_type='natural earth',  
        showland=True,  
        landcolor="LightGray",  
        showocean=True,  
        oceancolor="#DEEFF5" 
    )
)


fig.show()
```
