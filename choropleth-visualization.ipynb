```python
import pandas as pd 
df = pd.read_csv("/kaggle/input/dataset/owid-co2-data.csv")
```

```python
dff = df[["year","country","co2_per_capita","methane_per_capita","nitrous_oxide_per_capita"]]
dff.head()
```

```python
df1 = dff[(dff['year'] >= 1989) & (df['year'] <= 2023)]
df1 = df1[(df1["co2_per_capita"]>0) & (df1["methane_per_capita"]>0) & (df1["nitrous_oxide_per_capita"]>0)]
df1.head(50)
```

```python
df1.to_csv('per_capita_data.csv')
```

```python
import pandas as pd
import plotly.express as px

file_path = '/kaggle/working/per_capita_data.csv'  
df = pd.read_csv(file_path)


df_filtered = df[(df['year'] >= 1989) & (df['year'] <= 2023)]

color_min = df_filtered['co2_per_capita'].quantile(0.05)  
color_max = df_filtered['co2_per_capita'].quantile(0.95)  

fig = px.choropleth(
    df_filtered,
    locations='country',  # Column containing country names
    locationmode='country names',  # Set the location mode to country names
    color='co2_per_capita',  
    color_continuous_scale='oranges',  
    range_color=(color_min, color_max),  
    animation_frame='year',  # Column containing the years
    title='Choropleth Map of CO2 Per Capita Emissions Of All Countries (1990-2023)',
    hover_name='country',  
    projection='natural earth'  
)

fig.update_layout(
    width=1200,  
    height=800,  
    margin=dict(l=0, r=0, t=50, b=0),  # Adjust margins
    title_font_size=24,  # Set title font size
    title_x=0.5,  # Center the title
    geo=dict(
        showframe=False,  
        showcoastlines=True,  
        coastlinecolor="Black",  
        projection_type='natural earth',  
        showland=True,  
        landcolor="LightGray",  
        showocean=True,  
        oceancolor="#DEEFF5" 
    )
)


fig.show()
```
